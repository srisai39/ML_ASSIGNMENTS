```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris
```

```python
X=iris.data
Y=iris.target
```

```python
#reducing into 3D using PCA()
X1=decomposition.PCA(n_components=3).fit_transform(X)
```

```python
X1
```

# direct plotting

```python
fig=plt.figure(1,figsize=(8,8))
ax=Axes3D(fig,elev=50,azim=134)

ax.set_xlabel('principal component1')
ax.set_ylabel('principal component2')
ax.set_zlabel('principal component3')
ax.set_title('3D Graph for iris dataset')
target=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
ax.scatter(X1[:,0],X1[:,1],X1[:,2],c=Y,cmap=plt.cm.Set1,edgecolor='k',s=70)



ax.grid()
plt.show()
```
