```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# data

```python
temp_max=np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min=np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
months=np.arange(12)
```

```python
#plt.plot(months,temp_max,'ro')
#plt.plot(months,temp_min,'bo')
```

# fitting to periodic function using model func

```python
from scipy import optimize
```

```python
def model(time,avg,amp,time_offset):
    return (avg+amp*np.cos((time+time_offset)*2*np.pi/time.max()))
res_max,cov_max=optimize.curve_fit(model,months,temp_max,p0=[20,10,0])
res_min,cov_min=optimize.curve_fit(model,months,temp_min,p0=[-40,20,0])
```

```python
#res_max
```

# plot the fit

```python
days=np.linspace(0,12,num=365)
```

```python
plt.figure(figsize=(8,8))

plt.plot(months,temp_max,'ro')
plt.plot(days,model(days,*res_max),'r-')

plt.plot(months,temp_min,'bo')
plt.plot(days,model(days,*res_min),'b-')

plt.xlabel('Months')
plt.ylabel('Temperature(in C)')

plt.show()
```
